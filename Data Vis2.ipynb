```python
data = load_iris(as_frame = True).frame
data.head()

x = data['target'].value_counts().keys()
height = data['target'].value_counts().values

plt.bar(x,height)
plt.title('Bar Chart', color = 'Red', fontweight = 'bold')
plt.xticks(x)
plt.xlabel('X - Axis', color = 'Green')
plt.ylabel('Y - Axis' , color = 'Green')

plt.show()
```

```python
x = data.select_dtypes(float)
x.dtypes
```

```python
x.hist()
plt.show()
```

```python
plt.hist(x['sepal length (cm)'])
print(x['sepal length (cm)'].skew())
plt.title('Histogram', color = 'Purple', fontweight = 'bold', fontsize = 14)
plt.show()
```

```python
fig , axes = plt.subplots(2,2)

axes[0][0].hist(x['petal length (cm)'])
axes[0][0].set_title('petal length (cm)', color = 'Purple')

axes[0][1].hist(x['sepal length (cm)'])
axes[0][1].set_title('sepal length (cm)', color = 'Purple')

axes[1][0].hist(x['petal width (cm)'])
axes[1][0].set_title('petal width (cm)', color = 'Purple')

axes[1][1].hist(x['sepal width (cm)'])
axes[1][1].set_title('sepal width (cm)', color = 'Purple')

plt.tight_layout()
plt.show()
```

```python
#plt.subplot(nrows,ncols,pos)

plt.figure()

for i in range(len(x.columns)):
    plt.subplot(5,2,i+1)
    plt.hist(x[x.columns[i]])
    plt.title(x.columns[i]) 

plt.tight_layout()
plt.show()
```
