```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors

# Define a list of color names from matplotlib and create a color palette
color_list = list(colors.cnames)
np.random.seed(0)  # For reproducibility
color_palette = np.random.choice(color_list, size=2)

# Fetch historical data for EUR/USD and USD/JPY
eur = yf.Ticker("EURUSD=X")
usd_jpy = yf.Ticker("JPY=X")

# Get historical data
data = eur.history()
jpy_data = usd_jpy.history()

# Plot EUR/USD data
plt.figure(figsize=(14, 6))
plt.plot(data.index, data['Close'], color=color_palette[0], linestyle='-', linewidth=2, label='EUR/USD')
plt.title('EURO/USD Exchange Rate', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Date', fontsize=14, fontweight='bold')
plt.ylabel('Close Price (USD)', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='best')
plt.tight_layout()
plt.show()

# Plot USD/JPY data
plt.figure(figsize=(14, 6))
plt.plot(jpy_data.index, jpy_data['Close'], color=color_palette[1], linestyle='--', linewidth=2, label='USD/JPY')
plt.title('USD/JPY Exchange Rate', fontsize=16, fontweight='bold', color='darkgreen')
plt.xlabel('Date', fontsize=14, fontweight='bold')
plt.ylabel('Close Price (JPY)', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
